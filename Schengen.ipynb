```python
import pandas as pd
from datetime import timedelta, date, datetime
```

```python
df = pd.read_csv("days.csv")
```

```python
df = df[df.columns[:4]].head(10)
```

```python
df['Entry'] = pd.to_datetime(df['Entry'])
df['Exit'] = pd.to_datetime(df['Exit'])
```

```python
df
```

```python
df['number_of_days'] = df['Exit'] - df['Entry'] + timedelta(days=1)
df
```

```python
def days_within_180(df, por='latest'):
    if por == 'latest':
        por = df.iloc[-1]['Exit']
    if por == 'today':
        por = datetime.now().replace(hour=0, minute=0, second=0, microsecond=0)
    daysago180 = por - timedelta(days=180)
        
    cnt = timedelta(days=0)
    for entry, exit in zip(df['Entry'], df['Exit']):
        if exit < daysago180:
            cnt +=  timedelta(days=0)
        elif entry > por:
            cnt +=  timedelta(days=0)
        else:
            if entry < daysago180: entry = daysago180
            if exit > por: exit = por
            number_of_days =  exit - entry + timedelta(days=1)  
            cnt += number_of_days            
    return cnt
```

```python
dw180 = []

for i in range(df.shape[0]):
    dw180.append(days_within_180(df.head(i+1)))
    
df['days_within_180'] = dw180
```
